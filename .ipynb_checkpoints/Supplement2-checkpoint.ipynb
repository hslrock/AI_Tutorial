```python
import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt


torch.manual_seed(0)
x=torch.tensor((3.,4.,6.))
y=torch.tensor((-2.))
loss_list=[]
pred_list=[]


class Model(nn.Module):
    def __init__(self,):
        super(Model,self).__init__()
        self.fc1=nn.Linear(3,2,bias=False)
        self.fc2=nn.Linear(2,1,bias=False)
    def forward(self,x):
        out = self.fc2(self.fc1(x))
        return out



model=Model()
loss=nn.MSELoss()
optimizer=optim.SGD(model.parameters(),lr=0.002)
first=True
for i in range(100):
    weight_list=np.zeros((8,1))
    count=0
    for parameter in model.parameters(): 
        
        for layer in parameter:
            for weight in layer:
                weight_list[count]=weight.detach()
                count+=1

    if first:
        weight_history=weight_list
        first=False
    else:
        weight_history=np.concatenate((weight_history,weight_list),1)
        
        
    pred=model(x)
    optimizer.zero_grad()
    loss_val=loss(pred,y)
    loss_list.append(loss_val.item())
    pred_list.append(pred.item())
    loss_val.backward()
    optimizer.step()
```

```python
x=list(np.arange(0,100,1))
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('loss and predicted_value')
ax1.plot(x, loss_list)
ax2.plot(x, pred_list)
ax2.plot(x, np.ones(100)*-2)
fig, (ax1, ax2, ax3,ax4,ax5,ax6) = plt.subplots(6)
fig.suptitle('weights update')
ax1.plot(x, weight_history[0])
ax2.plot(x, weight_history[1])
ax3.plot(x, weight_history[2])
ax4.plot(x, weight_history[3])
ax5.plot(x, weight_history[4])
ax6.plot(x, weight_history[5])

fig, (ax7,ax8) = plt.subplots(2)
fig.suptitle('weights update')
ax7.plot(x, weight_history[6])
ax8.plot(x, weight_history[7])
```

```python


import torch
import torch.nn as nn
inp = torch.tensor([1.0, 2.0, 3, 4, 5])

outplace_dropout = nn.Dropout(p=0.4)

print("input: ",inp)
output = outplace_dropout(inp)
print("output: ",output)
print("input: ",inp) # Notice that the input doesn't get changed here

inplace_droput = nn.Dropout(p=0.4, inplace=True)
print("input: ",inplace_droput(inp))

print("input: ",inp) # Notice that the input is changed now
```
